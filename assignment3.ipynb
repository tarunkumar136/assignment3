Explanation:
We loop through each number, check if it’s divisible by any smaller number. If not, it’s prime.




```python
numbers = [2, 3, 4, 5, 10, 13, 17, 20]

for num in numbers:
    if num > 1:  # prime numbers are greater than 1
        is_prime = True
        for i in range(2, num):  # check divisibility
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num)
```

Explanation:
We use conditionals: if divisible by 2 → count as even, else add to odd sum.


```python
def analyze_numbers(lst):
    even_count = 0
    odd_sum = 0
    for num in lst:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_sum += num
    return even_count, odd_sum

print(analyze_numbers([1, 2, 3, 4, 5, 6]))
```

Explanation:
Default values (rate=5, time=2) are used if not provided. You can call with positional or keyword arguments.


```python
def simple_interest(principal, rate=5, time=2):
    return (principal * rate * time) / 100

# Positional arguments
print(simple_interest(1000))

# Keyword arguments
print(simple_interest(principal=2000, rate=10, time=3))
```

Explanation:
We define a class with attributes and a method. The method calculates grade based on marks.




```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 50:
            return "C"
        else:
            return "D"

s1 = Student("Alice", 92)
s2 = Student("Bob", 67)

print(s1.name, s1.grade())
print(s2.name, s2.grade())
```

Explanation:
Balance is private (__balance). We use methods to safely deposit/withdraw without direct access.


```python
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance  # private attribute

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds")

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
account.withdraw(300)
print(account.get_balance())
```

Explanation:
We handle two exceptions: division by zero and invalid input type (like entering text instead of a number).


```python
try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    result = a / b
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Invalid input type.")
```

Explanation:
We create a custom exception class and raise it when age is outside valid range.


```python
class InvalidAgeException(Exception):
    pass

def check_age(age):
    if age < 0 or age > 120:
        raise InvalidAgeException("Age must be between 0 and 120.")
    else:
        print("Valid age:", age)

try:
    check_age(150)
except InvalidAgeException as e:
    print(e)
```

Explanation:
We read numbers from a file, calculate total and average. If file doesn’t exist, we catch the error.


```python
try:
    with open("numbers.txt", "r") as f:
        numbers = [int(num) for num in f.read().split()]
        total = sum(numbers)
        average = total / len(numbers)
        print("Total:", total)
        print("Average:", average)
except FileNotFoundError:
    print("Error: File does not exist.")
```
